**One hot encoder**

- one hot encoder as name says at a time one will be on and ohter will be off

- for example case_status has two labels 

    - Certified 
    
    - Denied
    
- when we apply one hot encoding on case_status , it creates two more extra column

    - case_status_Certified
    
    - case_status_Denied
    
|case_status|case_status_Certified|case_status_Denied|
|-|-|-|
|Certified|1|0|
|Denied|0|1|
|Denied|0|1|
|Denied|0|1|
|Certified|1|0|

**Advantages**

- when we develop ML model it is very important to have the columns to be independent of each other

- Here case_status creating 2 extra columns which are independent of each other

- which means the one row will have only one column as 1 value

**Disadvantage**

- if column has 100 unique labels , 100 new columns will be cretaed

- the data will become sparse means huge 

- the processing time will increase and memmory consumption will also increase

- this is called **Curse of diemensionality**
