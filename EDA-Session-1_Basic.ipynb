```python
name=['Ramesh', 'Suresh', 'Satish']
age=[20,25,30]
columns=['Name','Age']
index=['a','b','c']

# provide columns
data=zip(name, age)
pd.DataFrame(data,columns=columns,index=index)
```

```python
name=['Ramesh', 'Suresh', 'Satish']
age=[20,25,30]
data=zip(name, age)
columns=['Name','Age']
index=['A','B','C']
df =pd.DataFrame(data,columns=columns,index=index)
df
```

- if you wan to add new column 

- df['new column']

- you need to have the list which is having some elements

- elements need to have same number of elements as of df rows

- city_names['Hyd','Banglore','Channai']

```python
city_names=['Hyd','Blr','Chennai']
df['city']=city_names
df
```

**Step 6 Update existing columns**

- create a ew column and update the existing column is done is same way

```python
df['city']=['H','B','C']
df
```

**Step 7**
$drop$ $the$ $column$

- in order to drop the column, we need to use drop method
- it takes 3 parameters
    - drop column or row
    - mention the column name
    - axis
       - axis=1 reference as column
       - axis=0 reference as row
    - you want to create new dataframe out of it or
    - you want to ovwerride the existing dataframe 
        - inplace=true

```python
# dropping the column with name 'city'
df.drop('city'
        ,axis=1
       ,inplace=True)
df
```

```python
# droping the row with index 'C'
df.drop('C'
        ,axis=0
       ,inplace=True)
df
```
